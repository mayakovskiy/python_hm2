# Задача 61
1. Определить какое максимальное и минимальное
значение median_house_value
2. Показать максимальное median_house_value, где
median_income = 3.1250
3. Узнать какая максимальная population в зоне
минимального значения median_house_value

## Пояснение
1. Минимум population и строки, где population равен этому мнимому
2. Среди полученного, найти строки, где значение median_house_value ровняется мимому

```python
import pandas as pd

df = pd.read_csv('/content/sample_data/california_housing_test.csv')
```

```python
df["median_house_value"].max()
```

```python
min_value = df["median_house_value"].min()
min_value
```

```python
df[['median_house_value']][df["median_income"] == 3.1250].max()
```

```python
df[df["median_income"] == 3.1250]['median_house_value'].max()
```

```python
df[df['median_house_value']==df['median_house_value'].min()]['population'].max()
```

```python
df[['population']][df['median_house_value'] == min_value]
```

# ДЗ  - Задача 42

Узнать какая максимальная households в зоне минимального значения population.

```python
import pandas as pd

df = pd.read_csv('/content/sample_data/california_housing_train.csv')
```

```python
min_population = df['population'].min()
min_population
```

```python
df[df['population']==min_population]['households'].max()
```

```python
df[['households']][df['population'] == min_population]
```
